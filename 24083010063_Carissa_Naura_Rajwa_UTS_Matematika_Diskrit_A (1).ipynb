FEM    adalah    pendekatan    numerik    untuk memecahkan masalah perpindahan besar dan kontinum regangan besar yang berada di luar kemampuan metode elemen   hingga   klasik (Carbonell   et   al.,   2022). Diharapkan hasil penelitian ini    dapat    memberikan    sumbangsih    bagi    dunia geoteknik   khususnya   dan   menjadi   khazanah   baru dengan menggunakan metode elemen hingga LISA.

```python
# Analisis Konsonan dan Vokal dari Abstrak dari Artikel

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

abstrak = """
Wilayah IKN didominasi oleh daerah perbukitan dan dataran yang luas,
Melihat karakteristik tanah yang beragam dan elevasi tanah yang bervariasi,
sering terjadi bencana seperti tanah longsor dan penurunan muka tanah.
Penelitian ini melakukan validasi hasil LISA terhadap program analisis geoteknik
yang sangat sering digunakan dan khusus untuk perangkat lunak geoteknik.
Diharapkan hasil penelitian ini dapat memberikan sumbangsih bagi dunia geoteknik
khususnya menjadi khazanah baru dengan menggunakan metode elemen hingga LISA V.8 FEA.
Untuk sudut penurunan yang terjadi adalah 0.0623 meter, dimana pada analisis
menggunakan software geoteknik Geostudio Sigma/W , penurunan yang terjadi
pada titik yang sama pada tinjauan adalah 0.0633 meter, terdapat perbedaan
yang tidak signifikan dengan rasio 1.016 dari hasil software geoteknik
Geostudio Sigma/W dengan hasil LISA V.8 FEA.
"""
# data dibersihkan (agar tidak mengganggu perhitungan)
clean_text = re.sub(r'[^a-zA-Z\s]', '', abstrak)
words = clean_text.lower().split()

def hitung_vokal_konsonan(kata):
    vokal = 'aiueo'
    j_vokal = sum(1 for c in kata if c in vokal)
    j_konsonan = sum(1 for c in kata if c.isalpha() and c not in vokal)
    return j_vokal, j_konsonan

data = []
for kata in words:
    v, k = hitung_vokal_konsonan(kata)
    data.append({'kata': kata, 'vokal': v, 'konsonan': k})

df = pd.DataFrame(data)

#tabell
pivot = df.pivot_table(index='vokal', columns='konsonan', values='kata', aggfunc='count', fill_value=0)

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')
plt.title("Heatmap Kombinasi Konsonan–Vokal (Abstrak Efendi, 2023)")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()

print("Tabel Pivot Konsonan–Vokal:")
print(pivot)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
from collections import Counter

abstrak = """
Wilayah IKN didominasi oleh daerah perbukitan dan dataran yang luas, Melihat karakteristik tanah yang beragam dan elevasi tanah yang bervariasi, sering terjadi bencana seperti tanah longsor dan penurunan muka tanah. Penelitian ini melakukan validasi hasil LISA terhadap program analisis geoteknik yang sangat sering digunakan dan khusus untuk perangkat lunak geoteknik. Diharapkan hasil penelitian ini dapat memberikan sumbangsih bagi dunia geoteknik khususnya menjadi khazanah baru dengan menggunakan metode elemen hingga LISA V8 FEA. Untuk sudut penurunan yang terjadi adalah 0,0623 meter, dimana pada analisis menggunakan software geoteknik Geostudio Sigma/W , penurunan yang terjadi pada titik yang sama pada tinjauan adalah 0,0633 meter, terdapat perbedaan yang tidak signifikan dengan rasio 1,016 dari hasil software geoteknik Geostudio Sigma/W dengan hasil LISA V8 FEA.
"""

# Pisah teks jadi paragraf-paragraf
paragraf_list = [p.strip() for p in abstrak.split('\n') if p.strip()]

# Fungsi bantu: ambil konsonan & vokal pertama dari kata
def extract_cv(word):
    vowels = 'aiueo'
    word = word.lower()
    first_consonant = next((ch for ch in word if ch not in vowels), None)
    first_vowel = next((ch for ch in word if ch in vowels), None)
    return first_consonant, first_vowel

# Loop tiap paragraf untuk buat tabel & heatmap
for i, paragraf in enumerate(paragraf_list, start=1):
    # Ambil semua kata alfabet
    words = re.findall(r'\b[a-zA-Z]+\b', paragraf.lower())

    # Hitung frekuensi tiap kata
    word_counts = Counter(words)

    # Ambil hanya kata unik (yang muncul sekali)
    unique_words = [w for w, c in word_counts.items() if c == 1]

    print(f"\n=== Paragraf {i} ===")

    if unique_words:
        # Buat tabel dari kata unik
        data = []
        for idx, word in enumerate(unique_words, start=1):
            c, v = extract_cv(word)
            data.append((idx, word, c, v))

        df_unique = pd.DataFrame(data, columns=["No", "Kata Unik", "Konsonan Pertama", "Vokal Pertama"])
        display(df_unique)  # tampil tabel

        # Buat kombinasi konsonan–vokal dari kata unik
        df_cv = df_unique[["Konsonan Pertama", "Vokal Pertama"]].dropna()

        # Hitung frekuensi kombinasi
        if not df_cv.empty:
            heatmap_data = df_cv.value_counts().unstack(fill_value=0)

            # Plot heatmap
            plt.figure(figsize=(8,6))
            sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu")
            plt.title(f"Heatmap Kata Unik Paragraf Abatrak ")
            plt.xlabel("Vokal Pertama")
            plt.ylabel("Konsonan Pertama")
            plt.show()
        else:
            print("Tidak ada data untuk heatmap di paragraf ini.")
    else:
        print("Tidak ada kata unik di paragraf ini.")
```

Heatmap tersebut menggambarkan pola distribusi huruf konsonan pertama dan vokal pertama dari kata-kata unik yang muncul hanya satu kali di paragraf pertama teks abstrak.

- Sumbu vertikal (Konsonan Pertama) menunjukkan huruf konsonan yang pertama kali muncul dalam kata unik (misalnya b, d, k, m, dll).

- Sumbu horizontal (Vokal Pertama) menunjukkan huruf vokal pertama yang muncul setelah konsonan tersebut (misalnya a, e, i, o, u).

- Nilai di setiap sel menunjukkan jumlah kata unik yang memiliki kombinasi konsonan-vokal tersebut.

Contoh: pada kombinasi (d, a) bernilai 4, berarti ada 4 kata unik yang dimulai dengan konsonan d dan vokal pertama a.

Heatmap menunjukkan bahwa dalam paragraf pertama abstrak, kata unik paling banyak diawali oleh konsonan m dengan vokal pertama e (pola “me–”), diikuti d–a dan d–i. Pola ini mengindikasikan kecenderungan penggunaan kata kerja berimbuhan khas Bahasa Indonesia formal, seperti melihat, dilakukan, atau dikaji, yang umum dalam teks ilmiah.

```python
# Analisis Konsonan dan Vokal dari Kesimpulan dari Artikel

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

kesimpulan = """
Penelitian analisis penurunan tanah ini dengan perangkat lunak metode elemen hingga LISA V.8 FEA memberikanhasil dalam analisis penurunan tanah dan memberikan perilaku yang sama dan mendekati hasil.
perangkat lunak analisis geoteknik, dengan nilai rasio 1.016, namun karena pemodelan hanya dilakukan pada satu  sampel,
peneliti  berharap  ada  penelitian  lebih lanjut  tentang  penggunaan  perangkat  lunak  LISA  V.8 FEA ini untuk beberapa sampel
sehingga perilaku yang terjadi dapat divalidasi dengan baik dan dapat menjadi pilihan  untuk  digunakan  analisis  geoteknik  dengan metode elemen hingga.
"""
# data dibersihkan (agar tidak mengganggu perhitungan)
clean_text = re.sub(r'[^a-zA-Z\s]', '', kesimpulan)
words = clean_text.lower().split()

def hitung_vokal_konsonan(kata):
    vokal = 'aiueo'
    j_vokal = sum(1 for c in kata if c in vokal)
    j_konsonan = sum(1 for c in kata if c.isalpha() and c not in vokal)
    return j_vokal, j_konsonan

data = []
for kata in words:
    v, k = hitung_vokal_konsonan(kata)
    data.append({'kata': kata, 'vokal': v, 'konsonan': k})

df = pd.DataFrame(data)

#tabell
pivot = df.pivot_table(index='vokal', columns='konsonan', values='kata', aggfunc='count', fill_value=0)

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')
plt.title("Heatmap Kombinasi Konsonan–Vokal (Kesimpulan Artikel)")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()

print("Tabel Pivot Konsonan–Vokal:")
print(pivot)
```

Berbeda dengan bagian pendahuluan yang banyak memakai kata panjang teknis, bagian kesimpulan cenderung lebih ringkas dan efisien. Ada peningkatan jumlah kata dengan 4 vokal dan 4–5 konsonan, menandakan beberapa kalimat penutup menggunakan istilah ilmiah, tapi tetap tidak berlebihan. Distribusi keseluruhan terlihat lebih tersebar, menandakan variasi bentuk kata lebih tinggi meskipun teksnya lebih singkat.

Kesimpulannya menggunakan pilihan kata yang padat makna, tidak terlalu teknis, dan mudah dipahami, sesuai fungsinya untuk menegaskan hasil penelitian.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
from collections import Counter

kesimpulan = """
Penelitian  analisis  penurunan  tanah  ini  dengan perangkat lunak metode elemen hingga LISA V8 FEA memberikanhasil dalam analisis penurunan tanah dan memberikan perilaku yang sama dan mendekati hasil.perangkat lunak analisis geoteknik, dengan nilai rasio 1,016, namun karena pemodelan hanya dilakukan pada satu  sampel, peneliti  berharap  ada  penelitian  lebih lanjut  tentang  penggunaan  perangkat  lunak  LISA  V8 FEA ini untuk beberapa sampel sehingga perilaku yang terjadi dapat divalidasi dengan baik dan dapat menjadi pilihan  untuk  digunakan  analisis  geoteknik  dengan metode elemen hingga.
"""

# Pisah teks jadi paragraf-paragraf
paragraf_list = [p.strip() for p in kesimpulan.split('\n') if p.strip()]

# Fungsi bantu: ambil konsonan & vokal pertama dari kata
def extract_cv(word):
    vowels = 'aiueo'
    word = word.lower()
    first_consonant = next((ch for ch in word if ch not in vowels), None)
    first_vowel = next((ch for ch in word if ch in vowels), None)
    return first_consonant, first_vowel

# Loop tiap paragraf untuk buat tabel & heatmap
for i, paragraf in enumerate(paragraf_list, start=1):
    # Ambil semua kata alfabet
    words = re.findall(r'\b[a-zA-Z]+\b', paragraf.lower())

    # Hitung frekuensi tiap kata
    word_counts = Counter(words)

    # Ambil hanya kata unik (yang muncul sekali)
    unique_words = [w for w, c in word_counts.items() if c == 1]

    print(f"\n=== Paragraf {i} ===")

    if unique_words:
        # Buat tabel dari kata unik
        data = []
        for idx, word in enumerate(unique_words, start=1):
            c, v = extract_cv(word)
            data.append((idx, word, c, v))

        df_unique = pd.DataFrame(data, columns=["No", "Kata Unik", "Konsonan Pertama", "Vokal Pertama"])
        display(df_unique)  # tampil tabel

        # Buat kombinasi konsonan–vokal dari kata unik
        df_cv = df_unique[["Konsonan Pertama", "Vokal Pertama"]].dropna()

        # Hitung frekuensi kombinasi
        if not df_cv.empty:
            heatmap_data = df_cv.value_counts().unstack(fill_value=0)

            # Plot heatmap
            plt.figure(figsize=(8,6))
            sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu")
            plt.title(f"Heatmap Kata Unik Kesimpulan")
            plt.xlabel("Vokal Pertama")
            plt.ylabel("Konsonan Pertama")
            plt.show()
        else:
            print("Tidak ada data untuk heatmap di paragraf ini.")
    else:
        print("Tidak ada kata unik di paragraf ini.")
```

- Kombinasi dominan:
m–e (4 kata unik) menjadi kombinasi paling tinggi. Hal ini menunjukkan bahwa banyak kata dalam kesimpulan diawali dengan “me–”, seperti memberikan, metode, menggunakan, atau memvalidasi. Awalan ini mencerminkan verba aktif khas bahasa ilmiah yang menekankan tindakan atau hasil. d–i (3 kata unik) dan p–e (3 kata unik) juga menonjol, menggambarkan dominasi bentuk kata kerja pasif atau nominal ilmiah seperti dilakukan, dengan, penelitian, perangkat, atau penggunaan.

- Kombinasi menengah:
b–e, s–a, dan t–e masing-masing muncul 2 kali. Kata-kata dengan pola ini kemungkinan seperti berdasarkan, sampel, tanah, atau teknik, yang berfungsi memperkuat konteks teknis dalam penjelasan hasil penelitian.

- Kombinasi langka:
Sebagian besar kombinasi dengan vokal i atau konsonan jarang seperti r, n, k hanya muncul satu kali. Artinya, struktur kata di bagian kesimpulan lebih homogen dibandingkan bagian sebelumnya, dengan pengulangan awalan yang sama (terutama me– dan pe–).

Heatmap kesimpulan menunjukkan dominasi kombinasi huruf m–e, d–i, dan p–e, mencerminkan gaya bahasa ilmiah yang menekankan tindakan, hasil, dan instrumen penelitian. Pola ini memperlihatkan bahwa bagian kesimpulan menggunakan banyak kata kerja aktif seperti memberikan, menggunakan, dan memvalidasi, serta kata nominal seperti penelitian dan perangkat, yang sesuai dengan tujuan akhir teks ilmiah — menegaskan hasil dan menyarankan tindak lanjut penelitian.

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd
import re

abstrak = """
Wilayah IKN didominasi oleh daerah perbukitan dan dataran yang luas, Melihat karakteristik tanah yang beragam dan elevasi tanah yang bervariasi, sering terjadi bencana seperti tanah longsor dan penurunan muka tanah. Penelitian ini melakukan validasi hasil LISA terhadap program analisis geoteknik yang sangat sering digunakan dan khusus untuk perangkat lunak geoteknik. Diharapkan hasil penelitian ini dapat memberikan sumbangsih bagi dunia geoteknik khususnya menjadi khazanah baru dengan menggunakan metode elemen hingga LISA V8 FEA. Untuk sudut penurunan yang terjadi adalah 0,0623 meter, dimana pada analisis menggunakan software geoteknik Geostudio Sigma/W , penurunan yang terjadi pada titik yang sama pada tinjauan adalah 0,0633 meter, terdapat perbedaan yang tidak signifikan dengan rasio 1,016 dari hasil software geoteknik Geostudio Sigma/W dengan hasil LISA V8 FEA.
"""

factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Bersihkan teks dari tanda baca & ubah ke huruf kecil
clean_text = re.sub(r'[^a-zA-Z\s]', '', abstrak.lower())

# Pisahkan kata-kata
words = clean_text.split()

# Lakukan stemming
stems = [stemmer.stem(w) for w in words]

# Hitung frekuensi kata dasar
freq = pd.Series(stems).value_counts().reset_index()
freq.columns = ['kata_dasar', 'frekuensi']

print("10 Kata Dasar yang Paling Sering Muncul pada Abstrak:")
display(freq.head(10))
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd
import re

kesimpulan = """
Penelitian  analisis  penurunan  tanah  ini  dengan perangkat lunak metode elemen hingga LISA V8 FEA memberikanhasil dalam analisis penurunan tanah dan memberikan perilaku yang sama dan mendekati hasil.perangkat lunak analisis geoteknik, dengan nilai rasio 1,016, namun karena pemodelan hanya dilakukan pada satu  sampel, peneliti  berharap  ada  penelitian  lebih lanjut  tentang  penggunaan  perangkat  lunak  LISA  V8 FEA ini untuk beberapa sampel sehingga perilaku yang terjadi dapat divalidasi dengan baik dan dapat menjadi pilihan  untuk  digunakan  analisis  geoteknik  dengan metode elemen hingga.
"""

factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Bersihkan teks dari tanda baca & ubah ke huruf kecil
clean_text = re.sub(r'[^a-zA-Z\s]', '', kesimpulan.lower())

# Pisahkan kata-kata
words = clean_text.split()

# Lakukan stemming
stems = [stemmer.stem(w) for w in words]

# Hitung frekuensi kata dasar
freq = pd.Series(stems).value_counts().reset_index()
freq.columns = ['kata_dasar', 'frekuensi']

print("10 Kata Dasar yang Paling Sering Muncul pada Kesimpulan:")
display(freq.head(10))
```

Berdasarkan hasil stemming, kata yang paling sering muncul dalam bagian kesimpulan adalah “analisis”, “dengan”, “lunak”, “tanah”, “metode”, dan “elemen”.
Frekuensi tinggi kata “analisis” dan “lunak” menunjukkan bahwa kesimpulan berfokus pada hasil analisis menggunakan perangkat lunak berbasis metode elemen hingga (Finite Element Method/FEM), yaitu LISA V8 FEA.
Kata “tanah” dan “turun” menegaskan bahwa objek kajian penelitian ini adalah penurunan tanah (settlement), yang menjadi inti dari analisis geoteknik yang dilakukan.
Kemunculan kata “metode” dan “elemen” memperkuat bahwa penelitian ini mengandalkan pendekatan numerik FEM sebagai metode utama untuk melakukan validasi hasil.

Selain itu, dari kalimat aslinya terlihat bahwa penelitian menghasilkan perbandingan antara hasil LISA dan perangkat lunak geoteknik lain dengan rasio 1,016, yang berarti perbedaan hasilnya sangat kecil dan dapat diterima secara teknis.
